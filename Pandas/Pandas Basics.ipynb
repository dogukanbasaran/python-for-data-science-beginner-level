### Adding New Column with assign() and insert()

```python
df_exams = pd.read_csv('datasets/StudentsPerformance.csv')
df_exams.head()
```

#### Adding a new column

##### assign()

When to use assign?

<li>Add multiple columns in a single line of code</li>
<li>When you need to overwrite the values of an existing columns (best practice)</li>

It returns a new object (a copy) with all the original columns in addition to the new ones.

```python
import numpy as np
```

```python
# create random numbers for our new score columns
score1 = np.random.randint(1, 101, size=1000)
score2 = np.random.randint(1, 101, size=1000)
```

```python
# create a series using the random numbers
series1 = pd.Series(score1, index=np.arange(0, 1000))
series2 = pd.Series(score2, index=np.arange(0, 1000))
```

```python
# using assign() to add multiple columns
df_exams.assign(score1=series1, score2=series2)
```

##### insert()

Inserts a new column at a specific position or index.
 Unlike assign() method, it will not copy original object, it overrides the original one.

```python
# using index() to add a column at a specific position
df_exams.insert(1, 'test', series1)
```

```python
df_exams
```

### The value_counts() method

#### Counting

```python
# len function
len(df_exams['gender'])
```

```python
# count() method
df_exams['gender'].count()
```

```python
# counting gender elements by category
df_exams['gender'].value_counts()
```

```python
# return the relative frequencies
df_exams['gender'].value_counts(normalize=True) # as proportion
```

```python
df_exams['gender'].value_counts(normalize=True) * 100 # as percentage
```

```python
# counting "parental level of education" elements by category
df_exams['parental level of education'].value_counts()
```

```python
# return the relative frequency and round to 2 decimals
df_exams['parental level of education'].value_counts(normalize=True).round(2)
```

### Sort a DataFrame with sort_values()

```python
# sort by one column
df_exams.sort_values(by='math score')
```

```python
# sort descending by one column
df_exams.sort_values(by='math score', ascending=False)
```

```python
# sort descending by multiple columns
df_exams.sort_values(by=['reading score', 'math score'])
```

```python
#sort descending by multiple columns and update dataframe
df_exams.sort_values(by=['reading score', 'math score'], ascending=False, inplace=True)
```

```python
df_exams
```

```python
# sort descending with a key function
df_exams.sort_values('race/ethnicity', ascending=True, key=lambda col:col.str.lower())
```

### The set_index() and sort_index() methods

```python
import random
```

#### Creating Index

```python
# creating non-repetitive values for the index
new_index = np.arange(0, 1000)
```

```python
# shuffling the indexes
random.shuffle(new_index)
```

```python
# creating a new column with new index
df_exams['new_index'] = new_index
df_exams
```

#### Set Index

```python
# set new_index column as index
df_exams.set_index('new_index', inplace=True)
df_exams
```

#### Sort Index

```python
# sorting dataframe by new index
df_exams.sort_index()
```

```python
# sorting descending dataframe by index
df_exams.sort_index(ascending=False, inplace=True)
```

### Filtering a DataFrame Based on Two or More Conditions

```python
# find Apple laptops that cost more than 2000 euros
(df_laptops["Company"] == "Apple") & (df_laptops["Price_euros"] > 2000)
```

#### & (and)

```python
# filter dataframe based on multiple conditions
df_laptops[(df_laptops["Company"] == "Apple") & (df_laptops["Price_euros"] > 2000)]
```

#### | (or)

```python
# find laptops Apple or Dell laptops
(df_laptops["Company"] == "Apple") | (df_laptops["Company"] == "Dell")
```

```python
# filter dataframe based on multiple conditions
df_laptops[(df_laptops["Company"] == "Apple") | (df_laptops["Company"] == "Dell")].value_counts("Company")
```

```python
# find laptops from Apple or Dell that cost over 2000 euros
((df_laptops["Company"] == "Apple") | (df_laptops["Company"] == "Dell")) & (df_laptops["Price_euros"]) > 2000
```

```python
# filter dataframe based on multiple conditions
df_laptops[((df_laptops["Company"] == "Apple") | (df_laptops["Company"] == "Dell")) & (df_laptops["Price_euros"] > 2000)]
```

### Creating a Conditional Column from More Than Two Choices np.select()

```python
# create an array based on multiple price tiers (+2 choices)

# create "conditions" and "values" variables
conditions = [
    df_laptops["Price_euros"] > 3000,
    (df_laptops["Price_euros"] > 2000) & (df_laptops["Price_euros"] <= 3000),
    (df_laptops["Price_euros"] > 800) & (df_laptops["Price_euros"] <= 2000),
    df_laptops["Price_euros"] <= 800
    
]

values = ["Too Expensive", "Expensive", "Affordable", "Cheap"]
```

```python
# set it to a new column
df_laptops["Price_tiers"] = np.select(conditions, values)
```

```python
# show dataframe
df_laptops.head()
```

```python
# count values inside "price_tier" column
df_laptops["Price_tiers"].value_counts()
```

#### Exercises

```python
# create an array based on multiple screen size tiers (+2 choices)
# Too Big > 16, Big > 14, Small > 12, Too Small < 12
# create "conditions" and "values" variables

conditions = [
    df_laptops["Inches"] > 16,
    (df_laptops["Inches"] > 14) & (df_laptops["Inches"] <= 16),
    (df_laptops["Inches"] > 12) & (df_laptops["Inches"] <= 14),
    df_laptops["Inches"] <= 12,
    
]
values = ["Too Big", "Big", "Small", "Too Small"]
```

```python
# set it to a new column
df_laptops["screen_size_tiers"] = np.select(conditions, values)
```

```python
# show dataframe
df_laptops.head()
```

```python
# count values inside "screen_size" column
df_laptops["screen_size_tiers"].value_counts()
```

### The .isin() Method

#### isin(): Single Filtering

```python
# select Apple or HP laptops
df_laptops["Company"].isin(["Apple", "HP"])
```

```python
# filter dataframes
df_laptops[df_laptops["Company"].isin(["Apple", "HP"])].value_counts("Company")
```

#### isin(): Multiple Filtering

```python
# find notebooks or ultrabooks from Apple or HP
filter1 = df_laptops["TypeName"].isin(["Notebook", "Ultrabook"])
filter2 = df_laptops["Company"].isin(["Apple", "HP"])
```

```python
# filter dataframe
df_laptops[filter1 & filter2]
```

### Finding Duplicate Rows with the .duplicated() Method

#### duplicated()

```python
# find duplicates in one column/series
df_laptops.duplicated("laptop_ID")
```

```python
# show elements in dataframe with duplicates in column 'laptop_ID'
df_laptops[df_laptops.duplicated("laptop_ID")]
```

```python
# duplicates in two or more columns
df_laptops.duplicated(["Product", "TypeName", "Inches"])
```

```python
duplicated = df_laptops.duplicated(["Product", "TypeName", "Inches"])
```

```python
# show all the values duplicated (except for the first - keep = "first")
df_laptops[duplicated].sort_values(["Product", "TypeName"])
```

#### Example

Task: find the cheapest and most expensive laptop of each company using the sort_values() and duplicated() methods (keep "first", "last", and False)

```python
# sort dataframe ascending by company and price
df_laptops = df_laptops.sort_values(["Company", "Price_euros"])
# cheapest first and most expensive last
df_laptops
```

```python
# check all value categories in "Company" column
df_laptops.value_counts("Company")
```

```python
# duplicated values in "Company" columns (default -> keep = "first")
duplicated_first = df_laptops.duplicated("Company", keep="first")
```

```python
# show dataframe with duplicated values in "Company" column
df_laptops[duplicated_first]

# keep = "first" (cheapest laptop per company)
# show dataframe with non-duplicated values in "Company" column
df_laptops[~duplicated_first][["Company", "Price_euros"]]

# check all value categories
df_laptops[~duplicated_first].value_counts("Company")
```

```python
# keep = "last"
duplicated_last = df_laptops.duplicated("Company", keep="last")
# keep = False
duplicated_false = df_laptops.duplicated("Company", keep=False)
```

```python
# keep = "last" (most expensive laptop per company)
# show dataframe with non-duplicated values in "Company" column
df_laptops[~duplicated_last][["Company", "Price_euros"]]
```

```python
# keep = False
# show dataframe with non-duplicated values in "Company" column
df_laptops[~duplicated_false]
```

```python
# all elements in company column are repeated at least once
df_laptops.value_counts("Company")
```

### Drop Duplicating Elements with the .drop_duplicates() Method

#### drop_duplicates()

```python
# drop duplicates in two or more columns
df_laptops.drop_duplicates(["Company"])[["Company", "Price_euros"]]
```

```python
# sort dataframe ascending by company and price
# cheapest first and most expensive last
df_laptops = df_laptops.sort_values(["Company", "Price_euros"])
```

```python
# cheapest: keep = "first"
df_laptops.drop_duplicates(["Company"], keep="first")[["Company", "Price_euros"]]
```

```python
df_laptops.drop_duplicates(["Company"], keep="last", inplace=True, ignore_index=True)
```

```python
# Arguments
# inplace: drop duplicates in place or to return a copy.
# ignore_index = True: the resulting axis will be labelled 0, 1, ....
df_laptops[["Company", "Price_euros"]]
```

### Getting and Counting Unique Values with the unique() and nunique() methods

#### unique()

Return unique values of Series object. (Uniques are returned in order of appearance.)

```python
# get unique elements in the "Company" column
df_laptops["Company"].unique()
```

```python
# get unique elements in the "Inches" column
df_laptops["Inches"].unique()
```

```python
# get the length of unique elements
len(df_laptops["Inches"].unique())
```

#### nunique()

Returns number of unique elements in the object (Excludes NA values by default)

```python
# get number of unique elements in the "Company" column
df_laptops["Company"].nunique()
```

```python
# get number of unique elements in the "Inches" column
df_laptops["Inches"].nunique()
```
